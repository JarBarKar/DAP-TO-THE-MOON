```python
def blob_polarity(text):
    return TextBlob(str(text)).sentiment.polarity

def blob_subjectivity(text):
    return TextBlob(str(text)).sentiment.subjectivity
```

```python
btc_headlines["subjectivity"]= btc_headlines["all_content_clean"].apply(blob_subjectivity)
btc_headlines["polarity"]= btc_headlines["all_content_clean"].apply(blob_polarity)
```

# Sentiment Analysis (using FinBert)

```python
!git clone https://gist.github.com/c1a8c0359fbde2f6dcb92065b8ffc5e3.git
```

```python
btc_headlines.head(5)
btc_headlines.dropna(inplace=True)
```

```python
import numpy as np

headlines_array = np.array(btc_headlines)
headlines_list= list(headlines_array[:, 5])
```

```python
!pip install transformers

#Getting the tokenizer and the model

from transformers import AutoTokenizer, AutoModelForSequenceClassification
  
tokenizer = AutoTokenizer.from_pretrained("ProsusAI/finbert")

model = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert")
```

```python
#That's where the headlines get tokenized to be inputted into model

inputs = tokenizer(headlines_list, padding = True, truncation = True, return_tensors='pt',is_split_into_words=False)
print(inputs)
```

```python
#inference

outputs = model(**inputs)
print(outputs.logits.shape)
```

```python
#Postprocessing with softmax

import torch

predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
print(predictions)
```

```python
#Formatting the results as a pandas data frame

import pandas as pd

#Headline #Positive #Negative #Neutral
positive = predictions[:, 0].tolist()
negative = predictions[:, 1].tolist()
neutral = predictions[:, 2].tolist()

table = {'Headline':headlines_list,
         "Positive":positive,
         "Negative":negative, 
         "Neutral":neutral}
      
df = pd.DataFrame(table, columns = ["Headline", "Positive", "Negative", "Neutral"])

df.head(5)
```
